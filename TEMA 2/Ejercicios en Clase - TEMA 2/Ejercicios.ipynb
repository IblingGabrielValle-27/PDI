```python
#ejercico 1.3

import cv2
import matplotlib.pyplot as plt
import numpy as np
plt.rcParams['figure.figsize']=[10,5]
img=cv2.imread('2.png',cv2.IMREAD_COLOR)
img_rgb=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)

r=img[:,:,2]
g=img[:,:,1]
b=img[:,:,0]

r_imagen=np.zeros_like(img_rgb)
r_imagen[:,:,0]=r

g_imagen=np.zeros_like(img_rgb)
g_imagen[:,:,1]=g

b_imagen=np.zeros_like(img_rgb)
b_imagen[:,:,2]=b

plt.subplot(141)
plt.title('Original')
plt.imshow(img_rgb)
plt.axis('off')

plt.subplot(142)
plt.title('Canal Rojo')
plt.imshow(r_imagen)
plt.axis('off')

plt.subplot(143)
plt.title('Canal Verde')
plt.imshow(g_imagen)
plt.axis('off')


plt.subplot(144)
plt.title('Canal Azul')
plt.imshow(b_imagen)
plt.axis('off')
```

```python
#ejercico 1.4
import cv2
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize']=[10,5]
img=cv2.imread('2.png',cv2.IMREAD_COLOR)
b,g,r= cv2.split(img)

plt.subplot(141)
plt.title('Original')
plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB))
plt.axis('off')

plt.subplot(142)
plt.title('Canal Rojo')
plt.imshow(r,cmap='Reds')
plt.axis('off')

plt.subplot(143)
plt.title('Canal Verde')
plt.imshow(g,cmap='Greens')
plt.axis('off')

plt.subplot(144)
plt.title('Canal Azul')
plt.imshow(b,cmap='Blues')
plt.axis('off')
plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

img_bgr= cv2.imread('3.jpeg')
img_rgb=cv2.cvtColor(img_bgr,cv2.COLOR_BGR2RGB)
print(f'Alto: {img_bgr.shape[0]}')
print(f'Ancho: {img_bgr.shape[1]}')
print(f'Canales: {img_bgr.shape[2]}')
fig,axs=plt.subplots(1,4,figsize=(20,5))

axs[0].imshow(img_rgb)
axs[0].set_title('imagen original')

fig_img=axs[1].imshow(img_rgb[...,0],cmap='gray')
fig.colorbar(fig_img, ax=axs[1], shrink=0.5)
axs[1].set_title('canal rojo')

fig_img = axs[2].imshow(img_rgb[..., 1], cmap='gray')
fig.colorbar(fig_img, ax=axs[2], shrink=0.5)
axs[2].set_title('Canal Verde')

fig_img = axs[3].imshow(img_rgb[..., 2], cmap='gray')
fig.colorbar(fig_img, ax=axs[3], shrink=0.5)
axs[3].set_title('Canal Azul')


```
